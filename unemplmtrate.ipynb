```python
import numpy as np
import pandas as pd
import calendar
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(path)
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed',
              'Estimated Labour Participation Rate', 'Region', 'longitude', 'latitude']
```

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
data['Frequency'] = data['Frequency'].astype('category')
data['Region'] = data['Region'].astype('category')
```

```python
data['Month'] = data['Date'].dt.month

data['Month_int'] = data['Month'].apply(lambda x: int(x))

data['Month_name'] = data['Month_int'].apply(lambda x: calendar.month_abbr[x])

data.drop('Month', axis = 1)
```

```python
data.head()
```

```python
data.info()
```

```python
data_stat = data[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']]
print(round(data_stat.describe(), 2))
```

```python
region_stat = data.groupby(['Region'])[['Estimated Unemployment Rate', 'Estimated Employed',
                                       'Estimated Labour Participation Rate']].mean().reset_index()
print(round(region_stat, 2))
```

DATA **VISUALIZATION**

```python
#  Estimated Unemployment Rate

from matplotlib import pyplot as plt
data['Estimated Unemployment Rate'].plot(kind='line', figsize=(8, 4), title='Estimated Unemployment Rate')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# Estimated Employed

from matplotlib import pyplot as plt
data['Estimated Employed'].plot(kind='line', figsize=(8, 4), title='Estimated Employed')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# Estimated Labour Participation Rate

from matplotlib import pyplot as plt
data['Estimated Labour Participation Rate'].plot(kind='line', figsize=(8, 4), title='Estimated Labour Participation Rate')
plt.gca().spines[['top', 'right']].set_visible(False)
```

HEATMAP

```python
hm = data[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'longitude', 'latitude', 'Month_int']]
hm = hm.corr()
plt.figure(figsize=(8,6))
sns.set_context('notebook', font_scale=1)
sns.heatmap(data=hm, annot=True, cmap="Purples")
```

BARPLOT

```python
import plotly.express as px
bar_p = data[['Estimated Unemployment Rate','States']]
data_unemployed = bar_p.groupby('States').mean().reset_index()

data_unemployed = data_unemployed.sort_values('Estimated Unemployment Rate')

fig = px.bar(data_unemployed, x='States',y='Estimated Unemployment Rate',color = 'States',title = 'Statewise average unemployment rate',
             template='ggplot2')
fig.show()
```

From the barplot we can see that the Unemployment rate is highest in 'Haryana' and lowest in 'Meghalaya'

SCATTERPLOT

```python
fig = px.scatter_matrix(data,template='plotly',dimensions=['Estimated Unemployment Rate', 'Estimated Employed',
                                                          'Estimated Labour Participation Rate'],color='Region')
fig.show()
```
